```python
#ODE

#2019331076

from sympy import *

from prettytable import PrettyTable

def function_val(function, x, y):
    return float(eval(eqn))

def euler(function, x, y, xf, step):
     
    yn = 0
    
    while( xf >= ( x + step ) ):
        
        yn = y + function_val ( function , x , y ) * step
        
        y = yn                                    ###Iterative Method
        
        x += step
        
    result = yn

    return result




def heun(function, x, y, xf, step):
    
    a1 = a2 = 1 / 2
    
    p = q = 1
    
    k1 = function_val ( function , x, y )
    
    k2 = function_val ( function , x + ( p * step ) , y + ( q * k1 * step ) )

    yn = y + ( a1 * k1 + a2 * k2 ) * step
    
    if ( xf == x + step ):
        
        result = yn
        
    else:
        
        result = heun ( function, x + step , yn , xf , step )
        
    return result


def midpoint( function , x , y , xf , step ):
    
    a1 = 0
    
    a2 = 1
    
    p = q = 1 / 2
    
    k1 = function_val( function , x, y )
    
    k2 = function_val(function , x + ( p * step ) , y + ( q * k1 * step ) )

    yn = y + ( a1 * k1 + a2 * k2 ) * step
    
    if ( xf == x + step ):
        
        result = yn
        
    else:
        
        result = midpoint ( function , x + step , yn , xf, step )
        
    return result

def ralston(function, x, y, xf, step):
    
    a1 = 1 / 3
    
    a2 = 2 / 3
    
    p = q = 3 / 4
    
    k1 = function_val( function , x , y )
    
    k2 = function_val( function , x + ( p * step ) , y + ( q * k1 * step ) )

    yn = y + (a1 * k1 + a2 * k2) * step
    
    if ( xf == x + step ):
        
        result = yn
        
    else:
        
        result = ralston( function , x + step , yn, xf, step)
        
    return result

eqn = input(" Enter your differential equation: ")

x0 = float(input(" x0 : "))

y0 = float(input(" y0 : "))

xf = float(input(" xf : "))

step = float(input(" Step size: "))

step_size = step

exact_solution = float(input(" Exact solution to the eqn: "))

ansTable = PrettyTable(['Step Size', 'Euler', 'Heun', 'Midpoint', 'Ralston'])

errorTable = PrettyTable(['Step Size', '|\N{GREEK SMALL LETTER EPSILON}|%(Euler)', '|\N{GREEK SMALL LETTER EPSILON}|%(Heun)', '|\N{GREEK SMALL LETTER EPSILON}|%(Midpoint)', '|\N{GREEK SMALL LETTER EPSILON}|%(Ralston)'])

cnt=0

while(cnt!=5):
    
    ansEuler = euler(eqn, x0, y0, xf, step)
    
    ansHeun = heun(eqn, x0, y0, xf, step)
    
    ansMidpoint = midpoint(eqn, x0, y0, xf, step)
    
    ansRalston = ralston(eqn, x0, y0, xf, step)

    ansTable.add_row([step, round(ansEuler,3), round(ansHeun,3), round(ansMidpoint,3), round(ansRalston,3)])
    
    errorTable.add_row([step, round((abs(exact_solution-ansEuler)/exact_solution)*100,3), round((abs(exact_solution-ansHeun)/exact_solution)*100,3),round((abs(exact_solution-ansMidpoint)/exact_solution)*100,3),round((abs(exact_solution-ansRalston)/exact_solution)*100,3)])

    cnt+=1
    
    step/=2

print("Comparison Table of different method of 'Ordinary Differential Equation' for step size: {}".format(step_size))

print(ansTable)

print("\n\n")

print("Error Table for comparison of different method of 'Ordinary Differential Equation' for step sizes: {}".format(step_size))

print(errorTable)























```
